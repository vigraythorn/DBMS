## Data generator for female customer info.

```python
from faker import Faker
import pandas as pd

maxnof = 500
maxnom = 500
fake = Faker()
## Customer Info. 
cid_ls=[]
fname_ls=[]
lname_ls=[]
add_ls=[]
email_ls=[]
country_ls=[]
gender_ls=[]
title_ls = []
phone_ls =[]

for i in range(1,maxnof+1):
    cid_ls.append(str(i))
    fname_ls.append(fake.first_name_female()) 
    lname_ls.append(fake.last_name_female()) 
    add_ls.append(fake.address())
    email_ls.append(fake.email())
    country_ls.append(fake.country())
    gender_ls.append('F') 
    title_ls.append('Mrs.') 
    phone_ls.append(fake.phone_number())

customer_info_f = list(zip(cid_ls, 
                           fname_ls,lname_ls,
                           add_ls,email_ls,
                           country_ls,gender_ls,
                           title_ls,phone_ls))

cust_info_f_df = pd.DataFrame(data=customer_info_f,
                              columns=['c_id',
                                       'cus_fname','cus_lname',
                                       'cus_addr','cus_email',
                                       'country','gender',
                                       'title','phone'])
#print(cust_info_f_df.tail(5))
```

```python
cid_ls=[]
fname_ls=[]
lname_ls=[]
add_ls=[]
email_ls=[]
country_ls=[]
gender_ls=[]
title_ls = []
phone_ls =[]

for i in range(maxnof+1,maxnof+maxnom+1):
    cid_ls.append(str(i))
    fname_ls.append(fake.first_name_male()) 
    lname_ls.append(fake.last_name_male()) 
    add_ls.append(fake.address())
    email_ls.append(fake.email())
    country_ls.append(fake.country())
    gender_ls.append('M') 
    title_ls.append('Mr.') 
    phone_ls.append(fake.phone_number())

customer_info_m = list(zip(cid_ls, 
                           fname_ls,lname_ls,
                           add_ls,email_ls,
                           country_ls,gender_ls,
                           title_ls,phone_ls))

cust_info_m_df = pd.DataFrame(data=customer_info_m,
                              columns=['c_id',
                                       'cus_fname','cus_lname',
                                       'cus_addr','cus_email',
                                       'country','gender',
                                       'title','phone'])
cust_info = pd.concat([cust_info_f_df,cust_info_m_df])
#print(cust_info.shape)
#print(cust_info.tail(3))
cust_info.to_csv('cust_info.csv',index=False,header=True)
```

## Booking

```python
from faker import Faker
import pandas as pd
fake = Faker()
import random
import time
def strTimeProp(start, end, format, prop):
    stime = time.mktime(time.strptime(start, format))
    etime = time.mktime(time.strptime(end, format))
    ptime = stime + prop * (etime - stime)
    return time.strftime(format, time.localtime(ptime))
def randomDate(start, end, prop):
    return strTimeProp(start, end, '%Y-%m-%d', prop) #%I:%M %p

bid_ls=[]
rm_id_ls=[]
bname_ls=[]
email_ls=[]
phone_ls =[]
#custno_ls=[]
b_sdate_ls=[]
nodays_ls=[]

for i in range(1,301):
    bid_ls.append(str(i))
    bname_ls.append(fake.name())
    email_ls.append(fake.email())
    phone_ls.append(fake.phone_number())
    #custno_ls.append(random.randint(1,5))
    b_sdate_ls.append(randomDate("2017-03-27", "2018-05-27", random.random()))
    nodays_ls.append(random.randint(1,3))
    
booking = list(zip(bid_ls,
                    bname_ls,email_ls,
                    phone_ls,      #custno_ls,
                    b_sdate_ls,nodays_ls))

booking_df = pd.DataFrame(data=booking,
                          columns=['b_id',
                                   'b_name','email',
                                   'phone',                #'cust_no',
                                  'bsdate','b_daysno'])
#print(booking_df.head(3))
#print(booking_df.shape)
booking_df.to_csv('booking.csv',header=True,index=False)
```
